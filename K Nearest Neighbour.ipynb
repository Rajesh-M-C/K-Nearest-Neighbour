```python
from sklearn.model_selection import train_test_split
```

```python
x = data.drop("diagnosis",axis=1)
y = data["diagnosis"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.3,shuffle=True,random_state=0)
```

```python
print("x_train",x_train.shape)
print("x_test",x_test.shape)
print("y_train",y_train.shape)
print("y_test",y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
for k in range(25):
    k_value = k+1
    knn = KNeighborsClassifier(n_neighbors=k_value,weights='uniform', algorithm='auto')
    knn.fit(x_train,y_train)
    y_predict=knn.predict(x_test)
    print("Accuracy:",round(accuracy_score(y_test,y_predict)*100,3),"%","K-value:",k_value)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn_grid = KNeighborsClassifier()
grid_search = GridSearchCV(knn_grid,param_grid=grid,cv=10,scoring="accuracy")
grid_search.fit(x_train,y_train)
print(grid_search.score)
```

```python
#Optimizing the number of neighbor using cross validation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
neighbors = []
cv_score = []
for k in range(1,51,2):
    neighbors.append(k)
    knn_c = KNeighborsClassifier(n_neighbors=k)
    knn_cv = cross_val_score(knn_c,x_train,y_train,cv=10,scoring="accuracy")
    cv_score.append(knn_cv.mean())
```

```python
# Misclassification error versus k
```

```python
mis_class_error = [1-x for x in cv_score]
```

```python
#determining the best k
```

```python
optimal_k = neighbors[mis_class_error.index(min(mis_class_error))]
print('The optimal number of neighbors is %d ' %optimal_k)
```

```python
aaa = pd.DataFrame(data=neighbors)
aaa["cv_score"] = cv_score
```

```python
aaa.sort_values(["cv_score"],ignore_index=True,ascending=False,inplace=True)
```
